```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/marketing_campaign.csv', sep = ';')
print('Количество жалоб:', df['Complain'].sum())
```

С помощью iloc берем срез по столбцам Mnt*. C помощью loc берем возраст покупателей.

```python
df['Year_Birth'] = df.Year_Birth.apply(lambda x: 2024-x)
customers_age = df.loc[df.iloc[0:,9:15].sum(axis=1) > 0, 'Year_Birth'] 
customers_avg_age = customers_age.mean()
avg_age = df['Year_Birth'].mean()
print('Средний возраст покупателей:', customers_avg_age)
print('Средний возраст всех людей:', avg_age)
```

Люди с базовым образованием имеют низкий доход и в среднем совершают меньше покупок. Одинокие люди совершают меньше покупок. Подавляющая часть покупателей имеет доход не выше 100000. Покупатели с доходом выше среднего в основном младше 60. Люди с 2 детьми совершают меньше покупок, чем остальные люди.

```python
cols = ['Education', 'Marital_Status', 'Income', 'Kidhome', 'Teenhome',
        'Recency', 'NumWebPurchases', 'NumStorePurchases' ,'Year_Birth']
for i in range(len(cols)):
    for j in range(i+1,len(cols)):
        plt.xlabel(cols[i], color = 'Red')
        plt.ylabel(cols[j], color = 'Red')
        plt.scatter(df[cols[i]],df[cols[j]])
        plt.show()
    
```
